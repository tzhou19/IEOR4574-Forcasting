```python
import numpy as np
import math
import matplotlib.pyplot as plt  # To visualize
import pandas as pd  # To read data
from scipy.stats import norm
from scipy.stats import poisson
```

### Question 1a

```python
x_axis = np.arange(-8, 8, 0.01)
plt.plot(x_axis, norm.pdf(x_axis, 0, math.sqrt(1)))
plt.title("Normal distribution with mean = 0, varaince = 1")
plt.show()
```

```python
x_axis = np.arange(-15, 15, 0.01)
plt.plot(x_axis, norm.pdf(x_axis, -5, math.sqrt(6)))
plt.title("Normal distribution with mean = -5, varaince = 6")
plt.show()
```

```python
x_axis = np.arange(-105, 105, 0.01)
plt.plot(x_axis, norm.pdf(x_axis, 100, math.sqrt(0.5)))
plt.title("Normal distribution with mean = 100, varaince = 0.5")
plt.show()
```

```python
t = np.arange(0, 100)
s1 = np.random.poisson(50,100)
s2 = np.random.poisson(50,100)
s3 = np.random.poisson(50,100)
s4 = np.random.poisson(50,100)
```

```python
fig, axs = plt.subplots(4, 1, figsize=(15, 15))
ax1 = plt.subplot(411)
plt.plot(t, s1)
plt.xlim(0, 100)
plt.ylim(20, 80)
plt.ylabel('$X_1$(t)',fontsize=20)
plt.tick_params('x', labelbottom=False)
plt.yticks(fontsize=12)
ax1.xaxis.grid(linewidth=2)
ax1.yaxis.grid(linewidth=1)

# share x only
ax2 = plt.subplot(412, sharex=ax1)
plt.plot(t, s2)
plt.xlim(0, 100)
plt.ylim(20, 80)
plt.ylabel('$X_2$(t)',fontsize=20)
# make these tick labels invisible
plt.tick_params('x', labelbottom=False)
plt.yticks(fontsize=12)
ax2.xaxis.grid(linewidth=2)
ax2.yaxis.grid(linewidth=1)

# share x only
ax3 = plt.subplot(413, sharex=ax1)
plt.plot(t, s3)
plt.xlim(0, 100)
plt.ylim(20, 80)
plt.ylabel('$X_3$(t)',fontsize=20)
# make these tick labels invisible
plt.tick_params('x', labelbottom=False)
plt.yticks(fontsize=12)
ax3.xaxis.grid(linewidth=2)
ax3.yaxis.grid(linewidth=1)


# share x and y
ax4 = plt.subplot(414, sharex=ax1, sharey=ax1)
plt.plot(t, s4)
plt.xlim(0, 100)
plt.ylim(20, 80)
plt.ylabel('$X_4$(t)',fontsize=20)
ax4.xaxis.grid(linewidth=2)
ax4.yaxis.grid(linewidth=1)
plt.xticks(fontsize=20)
# plt.savefig('Stochastic_Processes.png')
fig.suptitle("Poisson Stochastic Process", fontsize=20)
plt.show()
```
