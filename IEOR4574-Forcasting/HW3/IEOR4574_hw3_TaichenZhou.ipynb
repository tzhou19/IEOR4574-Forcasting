## AR

```python
def myAutoReg(y,n,c,phi,mu,sigma):
    return c + np.random.normal(mu,sigma) + phi*y
```

```python
mu = 0
sigma = 2
c = 0
t = np.arange(100)
fig = plt.figure(figsize=(20, 14))

phi=0.01
y = np.zeros(100)
y[0] = np.random.normal(mu,sigma)
for i in np.arange(99)+1:
    y[i] = myAutoReg(y[i-1],i,c,phi,mu,sigma)
ax1 = plt.subplot(311)
plt.plot(t, y,'k',linewidth=3.5)
plt.xlim(0, 100)
plt.ylim(-10, 10)
plt.ylabel('$y$(t)',fontsize=25)
plt.tick_params('x', labelbottom=False)
plt.yticks(fontsize=12)
plt.title('AR(1), c=0, phi = 0.01')
ax1.xaxis.grid(linewidth=2)
ax1.yaxis.grid(linewidth=1)
ax1.legend(['$\phi=0.01$'],fontsize=20)

phi=0.5
y = np.zeros(100)
y[0] = np.random.normal(mu,sigma)
for i in np.arange(99)+1:
    y[i] = myAutoReg(y[i-1],i,c,phi,mu,sigma)
ax2 = plt.subplot(312)
plt.plot(t, y,'k',linewidth=3.5)
plt.xlim(0, 100)
plt.ylim(-10, 10)
plt.ylabel('$y$(t)',fontsize=25)
plt.tick_params('x', labelbottom=False)
plt.yticks(fontsize=12)
plt.title('AR(1), c=0, phi = 0.5')
ax2.xaxis.grid(linewidth=2)
ax2.yaxis.grid(linewidth=1)
ax2.legend(['$\phi=0.5$'],fontsize=20)

phi=0.9
y = np.zeros(100)
y[0] = np.random.normal(mu,sigma)
for i in np.arange(99)+1:
    y[i] = myAutoReg(y[i-1],i,c,phi,mu,sigma)
ax3 = plt.subplot(313)
plt.plot(t, y,'k',linewidth=3.5)
plt.xlim(0, 100)
plt.ylim(-10, 10)
plt.ylabel('$y$(t)',fontsize=25)
plt.xlabel('t',fontsize=25)
plt.tick_params('x', labelbottom=False)
plt.yticks(fontsize=12)
ax3.xaxis.grid(linewidth=2)
ax3.yaxis.grid(linewidth=1)
ax3.legend(['$\phi=0.9$'],fontsize=20)
plt.grid(True)
plt.title('AR(1), c=0, phi = 0.9')
# plt.savefig('AR_1_c0.png')
plt.show()
```

```python
def myAutoReg_2(y_1, y_2,n,c,phi_1,phi_2,mu,sigma):
    return c + np.random.normal(mu,sigma) + phi_1*y_1 + phi_2*y_2
```

```python
mu = 0
sigma = 2
c = 0
t = np.arange(100)
fig = plt.figure(figsize=(20, 14))

phi_1 = 0.01
phi_2 = 0.01

y = np.zeros(100)
y[0] = np.random.normal(mu,sigma)
for i in np.arange(99)+1:
    y[i] = myAutoReg_2(y[i-1],y[i-2],i,c,phi_1,phi_2,mu,sigma)
ax1 = plt.subplot(311)
plt.plot(t, y,'k',linewidth=3.5)
plt.xlim(0, 100)
plt.ylim(-10, 10)
plt.ylabel('$y$(t)',fontsize=25)
plt.tick_params('x', labelbottom=False)
plt.yticks(fontsize=12)
ax1.xaxis.grid(linewidth=2)
ax1.yaxis.grid(linewidth=1)
ax1.legend(['$\phi_1=0.01, \phi_2=0.01$'],fontsize=20)

phi_1 = 0.5
phi_2 = 0.5
y = np.zeros(100)
y[0] = np.random.normal(mu,sigma)
for i in np.arange(99)+1:
    y[i] = myAutoReg_2(y[i-1],y[i-2],i,c,phi_1,phi_2,mu,sigma)
ax2 = plt.subplot(312)
plt.plot(t, y,'k',linewidth=3.5)
plt.xlim(0, 100)
plt.ylim(-20, 20)
plt.ylabel('$y$(t)',fontsize=25)
plt.tick_params('x', labelbottom=False)
plt.yticks(fontsize=12)
ax2.xaxis.grid(linewidth=2)
ax2.yaxis.grid(linewidth=1)
ax2.legend(['$\phi_1=0.5, \phi_2=0.5$'],fontsize=20)

phi_1 = 0.9
phi_2 = 0.9
y = np.zeros(100)
y[0] = np.random.normal(mu,sigma)
for i in np.arange(99)+1:
    y[i] = myAutoReg_2(y[i-1],y[i-2],i,c,phi_1,phi_2,mu,sigma)
ax3 = plt.subplot(313)
plt.plot(t, y,'k',linewidth=3.5)
plt.xlim(0, 100)
plt.ylim(0, 3000)
plt.ylabel('$y$(t)',fontsize=25)
plt.xlabel('t',fontsize=25)
plt.tick_params('x', labelbottom=False)
plt.yticks(fontsize=12)
ax3.xaxis.grid(linewidth=2)
ax3.yaxis.grid(linewidth=1)
ax3.legend(['$\phi_1=0.9, \phi_2=0.9$'],fontsize=20)
plt.grid(True)
# plt.savefig('AR_1_c0.png')
plt.show()
```

```python
dicfn = pd.read_csv('IPN31152N_Fred_Industrial Production Manufacturing Non-Durable Goods Ice Cream and Frozen Dessert.csv')
dicfn.DATE = pd.to_datetime(dicfn.DATE)
t = [i%365 for i in range(0, len(dicfn.DATE[dicfn.DATE.dt.year > 2015]))]
sig = dicfn.IPN31152N[dicfn.DATE.dt.year > 2015].values
RMSE = 10000000000
fss=0
order=0
```

```python
for i in np.arange(3,20):
    for j in np.arange(i,30):
        sos = signal.butter(i, 1, 'lowpass', fs=j, output='sos')
        filtered = signal.sosfilt(sos, sig)
        freqs, psd = signal.welch(sig)
        freqs, psd = signal.welch(sig-filtered)
        MSE=100000000000000000000
        MSE = np.square(sig-filtered).mean() 
        if(RMSE > math.sqrt(MSE)):
            RMSE = math.sqrt(MSE)
            order = i
            fss = j
```

```python
sos = signal.butter(order, 1, 'lowpass', fs=fss, output='sos')
filtered = signal.sosfilt(sos, sig)
fig = plt.figure(figsize=(30, 24))
ax1 = plt.subplot(511)
plt.plot(t, sig,'k')
plt.xlabel('t',fontsize=25)
plt.legend(['Production Index'],fontsize=20)
freqs, psd = signal.welch(sig)
plt.grid(True)
ax2 = plt.subplot(512)
plt.plot(freqs, psd,'k')
plt.xlabel('freq',fontsize=25)
plt.legend(['PSD of Production Index'],fontsize=20)
plt.grid(True)
ax3 = plt.subplot(513)
plt.plot(t, filtered,'k')
plt.xlabel('t',fontsize=25)
plt.legend(['Production Index LP'],fontsize=20)
plt.grid(True)
ax4 = plt.subplot(514)
plt.plot(t,sig-filtered,'k')
plt.ylim(-25,25)
plt.xlabel('t',fontsize=25)
plt.legend(['Production Index - LP'],fontsize=20)
freqs, psd = signal.welch(sig-filtered)
plt.grid(True)
ax5 = plt.subplot(515)
plt.plot(freqs, psd,'k')
plt.xlabel('freq',fontsize=25)
plt.legend(['PSD of filtered Production Index'],fontsize=20)
plt.grid(True)
# plt.savefig('ProductionIndex_Butterworth.png')
plt.show()
MSE = np.square(sig-filtered).mean()
RMSE = math.sqrt(MSE)
```
